```python
import pandas as pd
```

```python
df = pd.read_csv(
    "2012-2022_FIA_WEC_FULL_LAP_DATA.csv",
    low_memory=False
)

filtered_df = df[
    (df['class'] == "HYPERCAR") &
    (df['season'] == "2021") &
    (df['team_no'].isin(["Alpine Elf Matmut #36", "Toyota Gazoo Racing #8","Toyota Gazoo Racing #7"]))].drop(columns=['Unnamed: 0','number'])

print(filtered_df)
filtered_df.to_csv("wec_hypercar_2021.csv", index=False)
```
